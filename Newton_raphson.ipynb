```python
#Metodo de Newton
import numpy as np
import matplotlib.pyplot as plt
def f(x):
 return x**2 - x - 2  #Función a solucionar
def fp(x):
 return 2*x-1 #Derivada de Función a solucionar
X=np.linspace(-5,5,100)  #Intervalo de graficación
Y=f(X)
plt.plot(X,Y)
plt.grid()
plt.show()
x0=-2 #Valor inicial
iteracion=0
tol=0.0001  #Tolerancia
itermax=1000  #Número de iteraciones máxima
x=x0
while abs(f(x))>tol and iteracion<itermax:
 x=x-f(x)/fp(x)
 iteracion+=1
 print("Iteración:",iteracion,"x=",x)
print("solucion: x=",x)
```
